```python
%matplotlib inline
import matplotlib.pylab as plt
import warnings
import time
from IPython import display
import numpy as np

y = np.array([8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68])
x = np.array([10., 8., 13., 9., 11., 14., 6., 4., 12., 7., 5.])
N = len(x)

# Design matrix
A = np.vstack((np.ones(N), x, x**2, x**3)).T

# Learning rate
eta = 0.0000001         
# initial parameters
w = np.array([2., 1.,0.1, 0.001])

def gdesc(eta, A ,w, y):
    beta = 0.99
    p = 0
    for epoch in range(1000):        
        err = y-A.dot(w)
        warnings.simplefilter("error", RuntimeWarning)
        try:
            E = np.sum(err**2)/N
        except RuntimeWarning:
            break
        dE = -2.*A.T.dot(err)/N
        p = dE + beta*p
        w = w - eta*p    
    return E

emins = list()
emins.append(0)

for i in range(50):
    if gdesc(eta,A,w,y) <= gdesc((eta + eta*0.1),A,w,y):
        eta = eta - eta*0.1
    else:
        eta = eta + eta*0.1
    emins.append(min(emins))    
    emins[i] = gdesc(eta,A,w,y)
    print(i,'E:',emins[i],"eta:",eta)
    
emins.remove(0)
print("Min E:", min(emins))
```
